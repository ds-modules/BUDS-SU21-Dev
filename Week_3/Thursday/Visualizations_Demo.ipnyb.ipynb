## NBA Data
Let's load in our NBA Salaries Data that we have previously worked with to practice. Run the following cell to read the appropriate file into a Table.

```python
nba = Table.read_table("nba_salaries.csv")
```

```python
nba
```

```python
nba.group("POSITION").sort("count", descending=True).barh("POSITION", "count")
```

```python
nba.group("POSITION", sum).sort("2015-2016 SALARY sum", descending=True).barh("POSITION", "2015-2016 SALARY sum")
```

```python
nba.group("TEAM", np.mean).sort("2015-2016 SALARY mean", descending=True).barh("TEAM", "2015-2016 SALARY mean")
```

## Census Data

```python
data = 'http://www2.census.gov/programs-surveys/popest/datasets/2010-2015/national/asrh/nc-est2015-agesex-res.csv'

full_census_table = Table.read_table(data)
partial_census_table = full_census_table.select('SEX', 'AGE', 'POPESTIMATE2010', 'POPESTIMATE2014')
us_pop = partial_census_table.relabeled('POPESTIMATE2010', '2010').relabeled('POPESTIMATE2014', '2014')
us_pop = us_pop.where("AGE", are.below(150)).sort('AGE')
```

```python
us_pop.group("AGE", np.mean).plot("AGE", "2010 mean")
```

```python
us_pop.group("AGE", np.mean).plot("AGE", "2010 mean")
```
