```python
diabetes = health_risk_factors[["Country", "Prevalence of diabetes  % of population ages 20 to 79 2019"]]

meat_consumption_2019 = meat_consumption[meat_consumption["TIME"] == 2019]
```

```python
code = pd.read_csv("countries_codes_and_coordinates.csv")
code = code[["Country", "Alpha-3 code"]]
code["Alpha-3 code"] = code["Alpha-3 code"].str.replace('"', '').astype(str)
code["Alpha-3 code"] = code["Alpha-3 code"].str.replace(' ', '').astype(str)


code
```

```python
meat_consumption_2019["LOCATION"] = meat_consumption_2019["LOCATION"].astype(str)
```

```python
merged = pd.merge(meat_consumption_2019, code, how = "left", left_on = "LOCATION", right_on = "Alpha-3 code")

fnl_merged = pd.merge(merged, diabetes, how = "left", left_on = "Country", right_on = "Country")

```

Meat consumption is measured in thousand tonnes of carcass weight (except for poultry expressed as ready to cook weight) and in kilograms of retail weight per capita. Carcass weight to retail weight conversion factors are: 0.7 for beef and veal, 0.78 for pigmeat, and 0.88 for both sheep meat and poultry meat.

```python



fnl_merged.loc[(fnl_merged["MEASURE"] == "THND_TONNE") & (fnl_merged["SUBJECT"] == "PIG")]["Value"] =fnl_merged.loc[(fnl_merged["MEASURE"] == "THND_TONNE") & (fnl_merged["SUBJECT"] == "PIG")]["Value"] *.78

fnl_merged.loc[(fnl_merged["MEASURE"] == "THND_TONNE") & (fnl_merged["SUBJECT"] == "BEEF")]["Value"] =fnl_merged.loc[(fnl_merged["MEASURE"] == "THND_TONNE") & (fnl_merged["SUBJECT"] == "BEEF")]["Value"] *.7

fnl_merged.loc[(fnl_merged["MEASURE"] == "THND_TONNE") & (fnl_merged["SUBJECT"] == "SHEEP")]["Value"] =fnl_merged.loc[(fnl_merged["MEASURE"] == "THND_TONNE") & (fnl_merged["SUBJECT"] == "SHEEP")]["Value"] *.88


fnl_merged.loc[(fnl_merged["MEASURE"] == "THND_TONNE") & (fnl_merged["SUBJECT"] == "POULTRY")]["Value"] =fnl_merged.loc[(fnl_merged["MEASURE"] == "THND_TONNE") & (fnl_merged["SUBJECT"] == "POULTRY")]["Value"] *.88

```

```python
fnl_merged = fnl_merged[fnl_merged["Value"] < 30000].dropna()
```

```python
fnl_merged.plot.scatter("Value", "Prevalence of diabetes  % of population ages 20 to 79 2019")
```

```python
pig = fnl_merged[fnl_merged["SUBJECT"] == "PIG"]


pig.plot.scatter("Value", "Prevalence of diabetes  % of population ages 20 to 79 2019")
```
