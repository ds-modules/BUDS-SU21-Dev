```python
responses = list(age_and_climate["CLIM9_W67"])

for i in range(len(responses)):
    if responses[i] == "A great deal":
        responses[i] = 1.0
    if responses[i] == "Some":
        responses[i] = 2.0
    if responses[i] == "Not too much":
        responses[i] = 3.0
    if responses[i] == "Not at all":
        responses[i] = 4.0
    if responses[i] == "Refused":
        responses[i] = np.nan
```

```python
age_and_climate["Response_Sentiment"] = responses
```

```python
age_and_climate
```

```python
age_and_climate = age_and_climate[["F_AGECAT", "Response_Sentiment"]]
age_and_climate
```

```python
new_age_and_climate = age_and_climate.groupby(["F_AGECAT"]).mean().reset_index()
```

```python
sns.catplot(x="F_AGECAT", y="Response_Sentiment",kind="bar", data=new_age_and_climate);
```

As a reminder, as sentiment increases, there is a growing belief that climate change is not affecting the nearby community
